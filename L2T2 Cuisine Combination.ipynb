### Statistical Summary

```python
cuisines = df1["Cuisines"]
```

```python
aggregate_ratings  =  df1["Aggregate rating"]
```

```python
cusine_combinations_counts =  cuisines.value_counts()
```

```python
most_common_cusines_combinations = cusine_combinations_counts.head(10)
print("Mot Common Cuisine Combinations : ")
print(most_common_cusines_combinations)
```

```python
cusine_combinations_abg_ratings = df1.groupby("Cuisines")["Aggregate rating"].mean()
```

```python
top_10_avg_ratings = cusine_combinations_abg_ratings.nlargest(10)
print("\nAverage ratings for the top 10 cuisine combinations : ")
print(top_10_avg_ratings)
```
