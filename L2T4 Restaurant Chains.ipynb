## Identify restaurant chains based on common names or identifiers

```python
chains = df1.groupby('Restaurant Name')
```

## Calculate average rating for each chain

```python
average_ratings = chains['Aggregate rating'].mean()
```

## Calculate popularity for each chain

```python
popularity = chains['Votes'].sum()
```

## Plotting

```python
plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
average_ratings.sort_values(ascending=False).plot(kind='bar', color='skyblue')
plt.title('Average Rating of Restaurant Chains')
plt.xlabel('Restaurant Chain')
plt.ylabel('Average Rating')
plt.tight_layout()
plt.show()
```

```python
# Plot popularity
plt.subplot(2, 1, 2)
popularity.sort_values(ascending=False).plot(kind='bar', color='salmon')
plt.title('Popularity of Restaurant Chains')
plt.xlabel('Restaurant Chain')
plt.ylabel('Popularity')

plt.tight_layout()
plt.show()
```
