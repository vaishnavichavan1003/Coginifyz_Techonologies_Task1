### Task : Plot the locations of restaurants on a map using longitude and latitude coordinates. Identify any atterns or clusters of restaurants in specific areas.

### Import Library

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

### Import Dataset

```python
df=pd.read_csv("Dataset .csv")
```

```python
# Top 5 records
df.head()
```

```python
# Bottom 5 records
df.tail()
```

```python
# Create Shallow copy
df1=df.copy()
```

### Information

```python
df1.info()
```

### Statistical Summary

```python
df1.describe().T
```

```python
# Plot using matplotlib and seaborn
```

```python
df2=df1[["Longitude","Latitude"]]
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df1, x='Longitude', y='Latitude', s=100)
plt.title('Restaurant Locations')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# Plot using folium
```

```python
import folium
```

```python
map = folium.Map(location=[40.7128, -74.0060], zoom_start=12)
for i, row in df2.iterrows():
    folium.Marker([row['Latitude'], row['Longitude']]).add_to(map)
map.save('restaurants_map.html')
```
